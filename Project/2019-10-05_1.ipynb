```python
import numpy as np
import pandas as pd 
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
from sklearn.preprocessing import scale 
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import roc_auc_score,roc_curve
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier  
```

```python
data = pd.read_csv('M30_EURUSD.csv')
data.head()
```

```python
# soru işaretleri olan satırlar veriden silindi.
data = data.replace("?", np.nan)
data = data.dropna()
data.head()
```

```python
data['30_signal'].value_counts()
data.drop(index = data[data['30_signal']=='NONE'].index, inplace = True)
```

```python
df = data.iloc[:,5:430] #datamızdan _signal kolonlarını attık
df.shape
```

```python
#categorical olan kolonlarımız
cat_column_names = ['ind_7','ind_11','ind_24','ind_38','ind_54','ind_57','ind_60','ind_63','ind_66','ind_69','ind_72','ind_75',
                    'ind_78','ind_81','ind_84','ind_87','ind_89','ind_91','ind_93','ind_95','ind_97','ind_99','ind_101',
                    'ind_103','ind_105','ind_107','ind_109', 'ind_111', 'ind_113', 'ind_115','ind_138','ind_141','ind_144',
                    'ind_157','ind_159','ind_161','ind_163','ind_165','ind_167','ind_169','ind_171','ind_173','ind_175',
                    'ind_177','ind_182','ind_184','ind_187','ind_190','ind_193','ind_196','ind_199','ind_202','ind_205',
                    'ind_208','ind_211','ind_213','ind_384','ind_386','ind_388','ind_390']
```

```python
# categorical kolonların dummy var. oalrak değiştirdik
dms = pd.get_dummies(df[cat_column_names])
dms_none_cols = dms.filter(regex = '_NONE').columns
for i in dms_none_cols:
    dms.drop(i,axis=1,inplace=True)
dms_red_cols = dms.filter(regex = '_RED').columns
for i in dms_red_cols:
    dms.drop(i,axis=1,inplace=True)

dms.head()
```

```python
#datadan categorical olan kolonları çıkarıyoruz ve type nı değiştiriyoruz
df = df.drop(cat_column_names,axis=1).astype("float64")
df.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
pca = PCA()
# pca = PCA(n_components= 5) n_components parametresi ile 5 yeni pca'e düşüreleceği bilgisini modele söyleyebiliriz. 
#-default olarak %100'e ulaşacak şekilde bulur.
```

```python
df_pca = pca.fit_transform(scale(df))
```

```python
arr = np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4)*100)
arr
```

```python
# tmp'den büyük değerleri değişken olarak aldık tmp'den küçük sayısı +1 tane sonra tmpden büyük olanı elde etmiş oluruz
tmp = 95
num_var = sum((arr < tmp)) + 1 
num_var
```

```python
df_reduced = pd.DataFrame(df_pca[:,0:num_var])
df_reduced.head()
```

```python
# dummy e çevirdiğimiz datalar ile pca sonrası numeric değerlerimizi birleştiriyoruz
X = pd.concat([df_reduced,dms],axis=1)
```

```python
#logistic regresyon için y değişkenimizi 0 1 olarak değiştiriyoruz
data_y = data['30_signal']
cleanup_nums = {'BUY':1,'SELL':0}
data_y.replace(cleanup_nums, inplace=True)
y = np.array(data_y)
y
```

```python
X.head()
```

```python
loj = sm.Logit(y, X)
loj_model = loj.fit()
loj_model.summary()
```

```python
from sklearn.linear_model import LogisticRegression
loj = LogisticRegression(solver = 'liblinear')
loj_model = loj.fit(X,y)
loj_model
```

```python
y_pred = loj_model.predict(X)
confusion_mat = confusion_matrix(y, y_pred)
accuracy = accuracy_score(y, y_pred)
print("Counfusion matrix: \n",confusion_mat)
print("Accuracy: ",accuracy)
```

```python
print(classification_report(y, y_pred))
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, shuffle = False)
```

```python
pca1 = PCA()
pca2 = PCA()
```

```python
X_train_pca = pca1.fit_transform(scale(X_train))
X_test_pca = pca2.fit_transform(scale(X_test))
```

```python
arr1 = np.cumsum(np.round(pca1.explained_variance_ratio_, decimals = 4)*100)
arr2 = np.cumsum(np.round(pca2.explained_variance_ratio_, decimals = 4)*100)
```

```python
# tmp'den büyük değerleri değişken olarak aldık tmp'den küçük sayısı +1 tane sonra tmpden büyük olanı elde etmiş oluruz
tmp = 0.95
num_var1 = sum((arr1 < tmp*100)) + 1 
num_var2 = sum((arr2 < tmp*100)) + 1 
print(num_var1, num_var2)
```

```python
X_train_reduced = pd.DataFrame(X_train_pca[:,0:num_var1])
X_test_reduced = pd.DataFrame(X_test_pca[:,0:num_var2])
```

```python
loj1 = LogisticRegression(solver = 'liblinear')
loj1_model = loj1.fit(X_train_reduced,y_train)
loj1_model
```

```python
y_pred_1 = loj1_model.predict(X_train_reduced)
confusion_mat = confusion_matrix(y_train, y_pred_1)
accuracy = accuracy_score(y_train, y_pred_1)
print("Counfusion matrix: \n",confusion_mat)
print("Accuracy: ",accuracy)
```

```python
print(classification_report(y_train, y_pred_1))
```

```python
loj2 = LogisticRegression(solver = 'liblinear')
loj2_model = loj2.fit(X_test_reduced,y_test)
loj2_model
```

```python
y_pred_2 = loj2_model.predict(X_test_reduced)
confusion_mat = confusion_matrix(y_test, y_pred_2)
accuracy = accuracy_score(y_test, y_pred_2)
print("Counfusion matrix: \n",confusion_mat)
print("Accuracy: ",accuracy)
```
